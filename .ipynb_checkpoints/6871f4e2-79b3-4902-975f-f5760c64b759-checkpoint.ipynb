## Корреляционный анализ

```python
ferma_main.info()
```

```python
phik_matrix = ferma_main.drop('id', axis=1).phik_matrix(interval_cols=['удой', 'эке', 'сырой_протеин', 'белок', 'жирность', 'спо'])
mask = np.triu(np.ones_like(phik_matrix, dtype=bool))
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(phik_matrix, annot=True, fmt=".2f", cmap="PuBuGn", mask=mask)
plt.title("Корреляционная матрица")
plt.show()
```

```python
#Список количественных признаков 
quantitative_signs = ['эке', 'сырой_протеин', 'спо', 'жирность', 'белок']

#Диаграммы рассеяния
plt.figure(figsize=(20, 10))

for i, feature in enumerate(quantitative_signs, 1):
    plt.subplot(2, 3, i)
    sns.scatterplot(data=ferma_main, x=feature, y='удой', hue='порода', palette=['royalblue', 'red'])
    plt.title(f'удой & {feature}')

plt.tight_layout()
plt.show()
```

Вывод:  
1. "эке" - между "удой" и "эке" нелинейная связь, об этом нам говорит график похожий на квадратичную параболу. Коэффициент коррелляции довольно высокий - 0.72.  
2. "сырой_протеин" - связь умеренная, но разброс значений слишком большой. Коэффициент коррелляции средний - 0.46.  
3. "спо" - наблюдается резкий рост значений после 0.91, при этом рост больше характерен для породы РефлешнСоверинг. Связь похожа на линейную, также довольно значительный коэффициент коррелляции - 0.66. Имеет смысл определить "спо" в две категории.
4. "жирность" - связь линейная, наиболее заметна на значениях 3.6-3.7 и увеличение удоя больше характерно для породы РефлешнСоверинг. Коэффициент коррелляции высокий - 0.71.  
5. "белок" - связь нелинейная, но наблюдается зависимость от породы: значения белка выше у породы Вик Бик Айдиал.   Коэффициент коррелляции низкий - 0.06.

Сейчас модель присваивает класс 1 неправильно в 12 случаях из 34. Чтобы увеличить точность модели, небходимо увеличивать precision выборки(наша цель - снизить количество ошибок первого рода, то есть добиться минимального значения FP). Это достигается путем регулирования порога классификации.

Выведем пять наиболеее высоких значений вероятности нашей текущей модели и сравним их с классами тестовой выборки.

```python
y_proba = clf.predict_proba(X_test)[:, 1]

proba_data = pd.DataFrame(zip(y_test, y_proba),
                          columns=['y_valid', 'y_proba']).sort_values(by='y_proba', ascending=False)

proba_data.head()
```

Теперь напишем функцию, которая покажет нам для каждого значения порога классификации значения таких параметров как 'precision', 'recall', 'FP', 'число предсказанных объектов', 'всего объектов класса 1'.

```python
def threshold_choice(threshold_list, proba_df):
    result_df = pd.DataFrame(columns=['порог', 'precision', 'recall', 'FP', 'число_предсказанных_объектов',
                                      'всего_объектов_класса_1'])

    for threshold in threshold_list:
        column_name = 'y_pred_' + str(threshold)

        # Precision
        precision = proba_df.loc[proba_df[column_name] == 1, 'y_valid'].mean()

        # Число предсказанных объектов
        predicted_objects = proba_df[column_name].sum()

        # Всего объектов класса 1
        total_objects_class_1 = proba_df['y_valid'].sum()

        # Recall
        recall = proba_df.loc[(proba_df[column_name] == 1) & (proba_df['y_valid'] == 1)].shape[
                     0] / total_objects_class_1

        # Количество ложноположительных предсказаний (FP)
        false_positives = proba_df.loc[(proba_df[column_name] == 1) & (proba_df['y_valid'] == 0)].shape[0]

        # Добавляем результаты в датафрейм
        result_df = result_df.append({
            'порог': threshold,
            'precision': precision,
            'recall': recall,
            'FP': false_positives,
            'число_предсказанных_объектов': predicted_objects,
            'всего_объектов_класса_1': total_objects_class_1,
        }, ignore_index=True)

    return result_df
```

Применим нашу функцию и выведем 5 наивысших значений предсказанной вероятности, а также соответствующие порогам классификации предсказанные классы.

```python
proba_data_1 = proba_data.copy(deep=True)

thresholds_1 = [round(i, 2) for i in np.linspace(0.1, 1, num=10, endpoint=False)]
columns = []

for i in thresholds_1:
    columns.append('y_pred_' + str(i))
    proba_data_1['y_pred_' + str(i)] = proba_data_1['y_proba'].apply(lambda x: 1 if x > i else 0)

proba_data_1.head(5)
```

```python
threshold_choice(thresholds_1, proba_data_1)
```

Промежуточный вывод: Значения ниже 0.64 дают нам большое количество FP, при этом значения выше 0.73 значительно влияют на метрику precision(соответственно, и recall). Рассмотрим интервал от 0.64 до 0.73 детальнее, повторив все те же действия. 

```python
# Более детальный перебор порогов
proba_data_2 = proba_data.copy(deep=True)

thresholds_2 = [round(i, 2) for i in np.linspace(0.64, 0.75, num=11, endpoint=False)]

columns = []

for i in thresholds_2:
    columns.append('y_pred_' + str(i))
    proba_data_2['y_pred_' + str(i)] = proba_data_2['y_proba'].apply(lambda x: 1 if x > i else 0)

proba_data_2.head(5)
```

```python
threshold_choice(thresholds_2, proba_data_2)
```

Промежуточный вывод: нижнее значение порога, при котором precision равно 1 - это 0.72. Метрики pecision и recall выглядят адекватно. На следующем шаге установим 0.72 как оптимальный порог классификации и сравним старые и новые метрики.

```python
%%time
# Выделяем оптимальный порог
OPTIMAL_THRESHOLD = 0.72

# Применение порога к предсказанным вероятностям
y_pred_threshold = (y_proba > OPTIMAL_THRESHOLD).astype(int)

clf_accuracy_new = accuracy_score(y_test, y_pred_threshold)
clf_recall_new = recall_score(y_test, y_pred_threshold)
clf_precision_new = precision_score(y_test, y_pred_threshold)

print(f'Old Accuracy: {accuracy}')
print(f'Old Recall: {recall_old}')
print(f'Old Precision: {precision_old}')
print('---')
print(f'New Accuracy: {clf_accuracy_new}')
print(f'New Recall: {clf_recall_new}')
print(f'New Precision: {clf_precision_new}')
print('---')
```
